```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_style('white')
```

#### Poster

```python
save_plots = False

train_log = '../data/logs/poster_log_train.txt'
test_log = '../data/logs/poster_log_test.txt'

# Import data
train_iters = []
train_acc = []
train_loss = []
train_learn_rates = []
test_iters = []
test_acc = []
test_loss = []

with open(train_log, 'r') as f:
    for line in f:
        iteration, acc, loss, lr = line.strip().split()
        train_iters.append(int(iteration))
        train_acc.append(float(acc))
        train_loss.append(float(loss))
        train_learn_rates.append(float(lr))

with open(test_log, 'r') as f:
    for line in f:
        iteration, acc, loss = line.strip().split()
        test_iters.append(int(iteration))
        test_acc.append(float(acc))
        test_loss.append(float(loss))

# Print important summmary information
print 'Number of training data: {}'.format(len(train_iters))
print 'Number of test data: {}'.format(len(test_iters))
print 'Training iterations: {}'.format(max(train_iters))
print '  Learning rate:'
print '    Starting: {}'.format(max(train_learn_rates))
print '    Ending: {}'.format(min(train_learn_rates))
print '  Accuracy:'
print '    Starting: {}'.format(train_acc[0])
print '    Min: {}'.format(min(train_acc))
print '    Ending: {}'.format(train_acc[-1])
print '    Max: {}'.format(max(train_acc))
print '  Loss:'
print '    Starting: {}'.format(train_loss[0])
print '    Max: {}'.format(max(train_loss))
print '    Ending: {}'.format(train_loss[-1])
print '    Min: {}'.format(min(train_loss))
print 'Test iterations: {}'.format(max(test_iters))
print '  Accuracy:'
print '    Starting: {}'.format(test_acc[0])
print '    Min: {}'.format(min(test_acc))
print '    Ending: {}'.format(test_acc[-1])
print '    Max: {}'.format(max(test_acc))
print '  Loss:'
print '    Starting: {}'.format(test_loss[0])
print '    Max: {}'.format(max(test_loss))
print '    Ending: {}'.format(test_loss[-1])
print '    Min: {}'.format(min(test_loss))


train_skip = 10
# Plot training/test accuracies
plt.figure()
plt.plot(train_iters[::train_skip], train_acc[::train_skip])
plt.plot(test_iters, test_acc)
plt.legend(['Training', 'Test'], loc='lower center')
plt.xlabel('Iterations', fontsize=14)
plt.ylabel('Classification accuracy', fontsize=14)
if save_plots:
    plt.savefig('../plots/poster_acc.eps', format='eps', dpi=1000)
    plt.savefig('../plots/poster_acc.png', format='png')
plt.show()

train_skip = 10
# Plot training/test losses
plt.figure()
plt.plot(train_iters[::train_skip], train_loss[::train_skip])
plt.plot(test_iters, test_loss)
plt.legend(['Training', 'Test'], loc='upper center')
plt.xlabel('Iterations', fontsize=14)
plt.ylabel('Loss', fontsize=14)
if save_plots:
    plt.savefig('../plots/poster_loss.eps', format='eps', dpi=1000)
    plt.savefig('../plots/poster_loss.png', format='png')
plt.show()
```
