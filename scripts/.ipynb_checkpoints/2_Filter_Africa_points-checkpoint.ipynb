First we need to download a shapefile of Africa and save it somewhere.

Ok, continent shapefiles are downloaded and saved in `~/scratch/GitHub/viirs-nightlights/data/shapefiles/`.

#### Exploring the shapefiles

```python
t0 = time.time()
# Loading shapefile driver
driver = ogr.GetDriverByName('ESRI Shapefile')
# Loading layer of all continents
continents_shp = '../data/shapefiles/continent.shp'
ds = driver.Open(continents_shp)
boundary_layer = ds.GetLayer(0)
n_continents = boundary_layer.GetFeatureCount()
print 'There are {} continents'.format(n_continents)

# Getting Africa feature
poly = boundary_layer.GetFeature(3)
africa_poly = poly.GetGeometryRef()
n_geometries = africa_poly.GetGeometryCount()
print 'There are {} geometries in Africa'.format(n_geometries)

# Find the geometry with the most points
max_points = 0
for i in xrange(n_geometries):
    africa_ring = africa_poly.GetGeometryRef(i)
    points = africa_ring.GetGeometryRef(0).GetPointCount()
    if points > max_points:
        max_points = points
        africa = africa_ring.GetGeometryRef(0)
print 'Max points: {}'.format(max_points)

# Make Africa Shapely polygon
ring_points = []
for i in xrange(africa.GetPointCount()):
    ring_point = africa.GetPoint_2D(i)
    ring_points.append(ring_point)
africa_polygon = Polygon(ring_points)
t1 = time.time()
print 'Took {} seconds total'.format(t1-t0)
africa_polygon
```

#### Pickle Africa polygon

```python
polygon_fn = '../data/shapefiles/africa.pkl'
pkl.dump(africa_polygon, open(polygon_fn, 'wb'))
```

#### Test function
Let's test the function `check_within_africa` in `neal_utils.py`.

```python
# Create grid
lats = np.linspace(-100, 100, 100)
lons = np.linspace(-180, 180, 100)
lons_grid, lats_grid = np.meshgrid(lons, lats)
```

```python
# Plot grid
plt.scatter(lons_grid.ravel(), lats_grid.ravel())
plt.show()
```

```python
# Filter for Africa-only points
t0 = time.time()
num_africa = 0
africa_lats = []
africa_lons = []
idx = 0
for i in xrange(lons_grid.shape[0]):
    for j in xrange(lons_grid.shape[1]):
        idx += 1
        lat = lats_grid[i,j]
        lon = lons_grid[i,j]
        if check_within_africa(lat, lon):
            num_africa += 1
            africa_lats.append(lat)
            africa_lons.append(lon)
        if idx % 100 == 0:
            t1 = time.time()
            print 'Checked {} points in {} seconds'.format(idx, t1-t0)
t2 = time.time()
print 'Checked {} points in {} seconds'.format(idx, t2-t0)
print 'Total points in Africa: {}'.format(num_africa)
```

```python
# Plot points in Africa
africa_locs = np.vstack((np.array(africa_lats), np.array(africa_lons))).T
plt.scatter(africa_locs[:,1], africa_locs[:,0])
plt.show()
```

Looks like the function works as expected!
